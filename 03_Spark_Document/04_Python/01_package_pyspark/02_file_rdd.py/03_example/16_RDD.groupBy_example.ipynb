```python
import findspark
findspark.init()
```

```python
import pyspark
sc = pyspark.SparkContext(appName="RDD.groupBy")
```

```python
rddA = sc.parallelize(range(0,10))
rddA.collect()
```

```python
rddB = rddA.groupBy(lambda x: x % 3)
rddB.collect()
```

```python
[(x, [t for t in y]) for (x, y) in rddB.collect()]
```

```python
sc.stop()
```
