```python
import findspark
findspark.init()
```

```python
import pyspark
```

```python
sc = pyspark.SparkContext(appName="sortBy")
```

```python
rdd_1 = sc.parallelize(["b", "a", "c"])
rdd_1.collect()
```

```python
rdd_2 = sc.parallelize([str(i+1) for i in range(2)])
rdd_2.collect()
```

```python
rdd_3 = rdd_1.cartesian(rdd_2)
rdd_3.collect()
```

```python
sort_1 = rdd_3.sortBy(lambda x: x[0])
sort_1.collect()
```

```python
sort_2 = rdd_3.sortBy(lambda x: (x[1], x[0]))
sort_2.collect()
```

```python
sc.stop()
```
