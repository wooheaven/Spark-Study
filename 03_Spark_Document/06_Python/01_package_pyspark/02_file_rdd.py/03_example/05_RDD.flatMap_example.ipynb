```python
import findspark
findspark.init()
```

```python
import pyspark
sc = pyspark.SparkContext(appName="RDD.flatMap")
```

```python
l = [1, 2, 3]
```

```python
for x in l:
    print(x, "in l")
    for y in range(1, x):
        print(x, "in l", y, "in range")
```

```python
rdd = sc.parallelize(l)
```

```python
rdd.collect()
```

```python
result = rdd.flatMap(lambda x: range(1, x))
```

```python
result.collect()
```

```python
compare = rdd.map(lambda x: range(1, x))
```

```python
compare.collect()
```

```python
compare_2 = rdd.map(lambda x: list(range(1,x)))
```

```python
compare_2.collect()
```

```python
sc.stop()
```
